```python
df.pivot_table(values='類流感健保就診人次',index=['年',"週",'就診類別'],columns='縣市',aggfunc='sum')   
# sum表示求總和, 也可使用mean求平均或len求計數
```

```python
df1 = df.pivot_table(values='類流感健保就診人次',index=['年',"週",'就診類別'],columns='縣市',aggfunc='sum')   
df2 = df1['台北市']
```

```python
# 求算台北市因流感住院及門診總人數
df3 = df2.groupby('就診類別').sum()
df3
```
